```python
# import required libraries
import sqlite3
import pandas as pd
from IPython.display import display

# set display options
pd.options.display.max_columns = 100 
pd.set_option("display.max_columns", None) 


# initiate a connection 
sql_conn = sqlite3.connect('test.db')

# prepare the query 
query = 'select * from Garfield;'

# declare & execute a cursor
crsr = sql_conn.cursor() 
crsr.execute(query)

# fetch all records and save it in a dataframe 
dft=pd.read_sql_query(query,sql_conn)

# checking rows n columns of dataframe 
dft_lst = dft.shape
dft_lst
dft.head(10)
```

```python
# drop index column from dft
dft=dft.drop(['index'], axis=1)
```

#--- Above we have imported  Garfield table from test DB into a dataframe called dft
#--- Below we will import the XLS info to the dataframe dfx

```python
# import Garfield Data from XL into a dataframe called dfx
dfx = pd.read_excel ('garfield_activity.xlsx')

# checking rows n columns of dataframe 
dfx_lst=dfx.shape 

dfx.head(10)
```

```python
# function to Find rows which are different between two DataFrames.
def dataframe_difference(src_df, tgt_df, which=None):
    
    # Casting the data types of both the dataframe before comparision
    src_df = src_df.applymap(lambda x: x.strip() if type(x) == str else x)
    tgt_df = tgt_df.applymap(lambda x: x.strip() if type(x) == str else x)

    # left df is source af and right df is target df
    comparison_df = src_df.merge(tgt_df, indicator=True, how='outer')
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] != 'both']
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]

    return diff_df
```

```python
# identifying if there are more rows/columns in any of the dataframes
if dfx_lst != dft_lst:
    print('rows/columns are different between the two')
else:
    print('same number of rows/columns are present')
```

```python
# find out which columns are different 
colt = dft.columns.to_list()
colx = dfx.columns.to_list()
for i in range(len(colx)):
    if (dft[colt[i]].dtypes != dfx[colx[i]].dtypes):
        print(colt[i], colx[i])
```

```python
# rows present only in XL
result = dataframe_difference(dfx,dft,'left_only')
display(result)
```

```python
# rows present only in table
result = dataframe_difference(dfx,dft,'right_only')
display(result)
```

```python
# rows present in XL & table both
result = dataframe_difference(dfx,dft,'both')
display(result)
```
