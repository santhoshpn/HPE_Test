```python
# import necessary libraries
import pandas as pd
import numpy as np

# read input XL file into a dataframe
data = pd.read_excel('dow_jones_index.xlsx', sheet_name='dow_jones_index')
data.head()
```

```python
# create a new column called data and assign the values from previous row to it

data['pre_open'] = data['open'].shift()
data.head()
```

```python
# as first row will not have previous row  - to get data from - seed the value to avoid calculation error
#data = data.fillna(value=15.82)

data.pre_open = np.where(data.pre_open.isnull(), data.open, data.pre_open)

data.head()
```

```python
# create a new column called loss_gain and assign the values of week over week % loss gain to it

data['loss_gain']= ( (data['open'] - data['pre_open'] )/ data['pre_open'])*100
data.head()
```

```python
# create a new dataframe called result and calculate mean(average) of week over week loss gain 

result = data.groupby('stock')['loss_gain'].mean().reset_index(name='average')
result.head()
```

```python
# display top n leaders by indices/tickers

n=10
leaders = result.sort_values(by='average',ascending=False)
leaders.head(n)
```

```python
# display top n laggers by indices/tickers

n=10
laggers = result.sort_values(by='average') 
laggers.head(n)
```
